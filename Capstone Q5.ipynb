```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'movieReplicationSet-2.csv')
lion = df['The Lion King (1994)'] 
child = df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)']
frames = [lion, child]
combined = pd.concat(frames,axis=1)
combined
```

```python
combined = combined[~np.isnan(combined).any(axis=1)] # row-wise removal
combined
```

```python
combined['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'].value_counts()
```

```python
df['Are you an only child? (1: Yes; 0: No; -1: Did not respond)'].value_counts()
```

```python
only = combined[combined['Are you an only child? (1: Yes; 0: No; -1: Did not respond)']==1]
only
```

```python
only.mean()
```

```python
siblings = combined[combined['Are you an only child? (1: Yes; 0: No; -1: Did not respond)']==0]
siblings
```

```python
siblings.mean()
```

```python
 # unequal variance
from scipy.stats import ttest_ind

ttest_ind(only, siblings, equal_var = False)
```
