```python
import pandas as pd
import numpy as np
df = pd.read_csv(r'movieReplicationSet-2.csv')
wolf = df['The Wolf of Wall Street (2013)'] 
viewingPref = df['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)']
frames = [wolf, viewingPref]
combined = pd.concat(frames,axis=1)
combined
```

```python
combined = combined[~np.isnan(combined).any(axis=1)] # row-wise removal
combined
```

```python
combined['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)'].value_counts()
```

```python
alone = combined[combined['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)']==1]
alone
```

```python
group = combined[combined['Movies are best enjoyed alone (1: Yes; 0: No; -1: Did not respond)']==0]
group
```

```python
alone.mean()
```

```python
group.mean()
```

```python
from scipy.stats import ttest_ind

ttest_ind(alone, group, equal_var = False) # p-val is the first val in the array we're looking for
```
